```python
cats = pd.read_html('https://en.wikipedia.org/wiki/List_of_cat_breeds', header=0)[1]
countries = pd.read_html('https://en.wikipedia.org/wiki/List_of_sovereign_states_and_dependent_territories_in_Asia', header=0)[1]
```

```python
for i, country in countries.iterrows():
  input_string = country[2]
  for pos, letter in enumerate(input_string):
    if letter.isupper() and letter.isalpha():
      if input_string[pos-1].islower() and letter.isalpha():
        if pos > 0:
          split_position = pos
          break
  first_part = input_string[:split_position]
  second_part = input_string[split_position:]
  country[2] = first_part
  countries.iloc[i,2] = country[2]
  print(country[2])
```

```python
for i, country in countries.iterrows():
  input_string = country[2]
  x = input_string.split("[")
  first_part = x[0]
  country[2] = first_part
  countries.iloc[i,2] = country[2]
  print(country[2])
```

```python
for i, cat in cats.iterrows():
  input_string = cat[1]
  x = input_string.split("[")
  first_part = x[0]
  cat[1] = first_part
  cats.iloc[i,1] = cat[1]

for i, cat in cats.iterrows():
  input_string = cat[1]
  x = input_string.split(";")
  first_part = x[0]
  cat[1] = first_part
  cats.iloc[i,1] = cat[1]

for i, cat in cats.iterrows():
  input_string = cat[1]
  x = input_string.split(",")
  first_part = x[0]
  cat[1] = first_part
  cats.iloc[i,1] = cat[1]

for i, cat in cats.iterrows():
  input_string = cat[1]
  x = input_string.split(" (")
  first_part = x[0]
  cat[1] = first_part
  cats.iloc[i,1] = cat[1]

for i, cat in cats.iterrows():
  input_string = cat[1]
  x = input_string.split("Developed in ")
  if(len(x) > 1):
    first_part = x[1]
    cat[1] = first_part
  cats.iloc[i,1] = cat[1]

for i, cat in cats.iterrows():
  input_string = cat[1]
  x = input_string.split(" in")
  first_part = x[0]
  cat[1] = first_part
  cats.iloc[i,1] = cat[1]

for i, cat in cats.iterrows():
  input_string = cat[1]
  x = input_string.split(" and")
  first_part = x[0]
  cat[1] = first_part
  cats.iloc[i,1] = cat[1]

for i, cat in cats.iterrows():
  input_string = cat[1]
  cat[1] = input_string.lstrip()
  cats.iloc[i,1] = cat[1]
  print(cat[1])

for i, cat in cats.iterrows():
  input_string = cat[0]
  x = input_string.split(" or")
  first_part = x[0]
  cat[0] = first_part
  cats.iloc[i,0] = cat[0]

for i, cat in cats.iterrows():
  input_string = cat[0]
  x = input_string.split(", ")
  first_part = x[0]
  cat[0] = first_part
  cats.iloc[i,0] = cat[0]

for i, cat in cats.iterrows():
  input_string = cat[0]
  x = input_string.split(" (")
  first_part = x[0]
  cat[0] = first_part
  cats.iloc[i,0] = cat[0]
```

```python
asianCats = []
currentCountry = countries.iloc[2]
for i, cat in cats.iterrows():
    currentCat = cat[1]
    for i, country in countries.iterrows():
        currentCountry = country[2]
        if(currentCat == currentCountry):
            asianCats.append({cat[0]:cat[1]})

asianCats
```

```python
url = "https://www.scrapethissite.com/pages/simple/"

response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser") 

```

```python
countriesArr = []
countries = soup.select(".country-name")
for i in countries:
    countriesArr.append(i.text)
for i in range(0, len(countriesArr)):
    countriesArr[i] = countriesArr[i].strip()
print(countriesArr)

capitalsArr = []
capitals = soup.select(".country-capital")
for i in capitals:
    capitalsArr.append(i.text)
print(capitalsArr)

populationArr = []
populations = soup.select(".country-population")
for i in populations:
    populationArr.append(i.text)
print(populationArr)

areaArr = []
areas = soup.select(".country-area")
for i in areas:
    areaArr.append(i.text)
print(areaArr)
```

```python
d = {"Country": countriesArr, "Capital": capitalsArr, "Population": populationArr, "Area":areaArr}
countryDf = pd.DataFrame(data=d)
```
